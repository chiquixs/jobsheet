```python
import pandas as pd
from scipy.stats import norm
from scipy import stats

# Mount google drive
drive.mount('/content/drive')

# Define the file path
file_path = '/content/drive/MyDrive/STATISTIKA/Titanic-Dataset.csv'

# Open the CSV file
try:
    data = pd.read_csv(file_path)
    print(data.head())
except Exception as e:
    print("Terjadi error:", e)

data['Age'] = data['Age'].fillna(data['Age'].mean())
print("Age is Imputation by mean:", round(data['Age'].mean(), 2))

print("total data with null value:", data['Age'].isnull().sum())


print("\n---Calculate z-score for 'Age' column---")

# calculate z-score
data['Z_Score_Age'] = stats.zscore(data['Age'])

print("Z-score for 'Age' column succes to calculated and will be save in 'Z_Score_Age' column")
print(data[['Age', 'Z_Score_Age']].head())

# Looking for Outlier
outliers = data[abs(data['Z_Score_Age']) > 3]

# Total Outlier
jumlah_outlier = outliers.shape[0]
print(f"Total outlier in 'Age' column is : {jumlah_outlier}")

# formula
mean_age = data['Age'].mean()
std_age = data['Age'].std()

# for probability < 20
peluang = norm.cdf(20, mean_age, std_age)
print(f"Probability (Age < 20) = {peluang:.4f}")




```
